# Data Collection

## Objective
The objective of this notebook is to collect SpaceX Falcon 9 launch data from multiple sources.

## Data Sources
- SpaceX REST API
- Wikipedia Falcon 9 launch records

## Description
The collected data includes information about:
- Launch site
- Orbit type
- Payload mass
- Launch outcome
- Landing outcome

## Tools Used
- Python
- Requests
- Pandas

## Summary
The collected dataset provides the foundation for further data wrangling, exploratory data analysis, and predictive modeling.
